```python
import numpy as np
import matplotlib.pyplot as plt

# Dados dos espectros para três estrelas
wavelengths1, intensities1 = np.loadtxt('s4000K.spec.nulbad.0.010', unpack=True)
wavelengths2, intensities2 = np.loadtxt("s6000K.spec.nulbad.0.010", unpack=True)
wavelengths3, intensities3 = np.loadtxt("s8000K.spec.nulbad.0.010", unpack=True)

# Criar subplots para visualizar os espectros separadamente
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True, figsize=(10, 12))

# Parâmetros de plot
line_colors = ['orangered', 'limegreen', 'royalblue']
line_labels = ['Estrela 4000K', 'Estrela 6000K', 'Estrela 8000K']
line_width = 0.8

# Plot para a primeira estrela
ax1.plot(wavelengths1, intensities1, color=line_colors[0], linewidth=line_width, label=line_labels[0])
ax1.set_xlabel("Comprimento de Onda (Å)")
ax1.set_ylabel('Fluxo')
ax1.set_title('Estrela 4000K')
ax1.legend()

# Plot para a segunda estrela
ax2.plot(wavelengths2, intensities2, color=line_colors[1], linewidth=line_width, label=line_labels[1])
ax2.set_xlabel("Comprimento de Onda (Å)")
ax2.set_ylabel('Fluxo')
ax2.set_title('Estrela 6000K')
ax2.legend()

# Plot para a terceira estrela
ax3.plot(wavelengths3, intensities3, color=line_colors[2], linewidth=line_width, label=line_labels[2])
ax3.set_xlabel("Comprimento de Onda (Å)")
ax3.set_ylabel('Fluxo')
ax3.set_title('Estrela 8000K')
ax3.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dados dos espectros para três estrelas
data_files = ['s4000K.spec.nulbad.0.010', 's6000K.spec.nulbad.0.010', 's8000K.spec.nulbad.0.010']

# Criar um único plot para visualizar os espectros juntos
plt.figure(figsize=(12, 6))

# Parâmetros das estrelas (substitua pelos valores reais)
parameters = {
    'Estrela 4000K': {'teff': 4000, 'glog': 4.4, 'asalog': 0.0, 'flprefix': 's4000K', 'llzero': 9200, 'llfin': 9250},
    'Estrela 6000K': {'teff': 6000, 'glog': 4.4, 'asalog': 0.0, 'flprefix': 's6000K', 'llzero': 9200, 'llfin': 9250},
    'Estrela 8000K': {'teff': 8000, 'glog': 4.4, 'asalog': 0.0, 'flprefix': 's8000K', 'llzero': 9200, 'llfin': 9250}
}

# Parâmetros de plot
line_colors = ['orangered', 'limegreen', 'royalblue']
line_labels = ['Estrela 4000K', 'Estrela 6000K', 'Estrela 8000K']
line_width = 0.8

for star_name, params in parameters.items():
    wavelengths, intensities = np.loadtxt(data_files[list(parameters.keys()).index(star_name)], unpack=True)
    label = f"{star_name.capitalize()} "

    normalized_intensities = intensities / np.max(intensities)  # Normalização
    plt.plot(wavelengths, normalized_intensities, color=line_colors[list(parameters.keys()).index(star_name)], linewidth=line_width, label=label)

plt.xlabel("Comprimento de Onda (Å)")
plt.ylabel("Fluxo")
plt.title("Espectros Normalizados das Três Estrelas")
plt.legend()
plt.show()
```
